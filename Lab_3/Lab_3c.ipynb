#### This dataset comprises 10,000 records, each represented by eight features and one target variable for machine failure. Every entry includes a unique identifier (UID) and a product ID that encodes the variant-specific serial number of the tool. Key operational parameters such as rotational speed (rpm), torque (Nm), and accumulated tool wear (minutes) are provided to reflect the mechanical load and degradation conditions.
